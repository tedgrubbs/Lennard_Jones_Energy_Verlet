```python
vDist = np.loadtxt("speeddistributionfile.csv")
fig = plt.figure(figsize=(7,5))
ax = fig.add_subplot(111)
ax.set_ylabel('P', fontsize='20')
ax.set_xlabel('V',fontsize='20')
ax.set_title('Velocity distribution', fontsize='20')
counts, bins, crap = plt.hist(vDist[:],50,density=True)
```

```python
correctX = np.zeros(len(counts))
for i in range(len(counts)):
    correctX[i] = (bins[i] + bins[i+1])/2
    
def f(v, T):
    return (v/T)*np.exp(-v**2/(2*T))

params, cov = curve_fit(f, correctX, counts)
    
fig = plt.figure(figsize=(7,5))
ax = fig.add_subplot(111)
ax.set_ylabel('P', fontsize='20')
ax.set_xlabel('V',fontsize='20')
ax.set_title('Velocity distribution', fontsize='20')
plt.plot(correctX, counts, 'o')
plt.plot(np.arange(0,4,.1), f(np.arange(0,4,.1), params[0]), label='MB Distribution of\nT=.795')
plt.legend(loc='upper right', fontsize=15)
print(params[0])
```
